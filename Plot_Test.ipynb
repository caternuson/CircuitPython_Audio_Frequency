```python
import matplotlib.pyplot as plt
```

```python
FILE_NAME = "recorder_c5.dat"
```

```python
import csv
x = []
y = []
with open(FILE_NAME) as f:
    reader = csv.reader(f)
    for r in reader:
        x.append(float(r[0]))
        y.append(float(r[1]))
```

```python
plt.figure()
plt.plot(x,y)
```
